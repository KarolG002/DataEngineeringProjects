### Let's start off with inspections dataset and do some EDA on it to find out what the data looks like and it's inconsistencies

```python
inspections.info()
no_rows = len(inspections)
print(f"Inspections df has {no_rows} rows")
```

```python
def missing_columns(df):
    total = 0
    for column in df.columns:
        missing = df[column].isnull().sum()
        total += missing
        if missing != 0:
            print(f"{column} --> {df[column].isnull().sum()}")
    if total == 0:
        print("No missing values")

missing_columns(inspections)
```

```python
null_ll_rows = inspections.query('Latitude != Latitude')
null_ll_rows.head()
```

#### As we can see where there are null values in Latitude and Longitude the same applies to few other columns so lets remove those rows and drop the column isSidewaycompliant, because its all null
